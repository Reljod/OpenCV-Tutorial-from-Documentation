# Smoothing Images
### Goals:
**Learn to:**
* Blur the images with various low pass filters
* Apply custom-made filters to images (2D convolution)

## 2D Convolution (Image Filtering)

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('images/ReljodImg.jpg')
kernel = np.ones((5,5),np.float32)/(5**2)
dst = cv.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Blurring
* Blurring removes high frequency content in an image like noise and edges.

## Averaging using cv.boxFilter

```python
img = cv.imread('images/ReljodImg.jpg')
kernel = np.ones((13,13),np.float32)/(13**2)
blur = cv.boxFilter(img, -1, (5,5), normalize=True)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Average using cv.blur and cv.GaussianBlur
**Gaussian Blur** 
 * it is used for removing Gaussian Noise in an image, but the edges are also smoothen out

```python
img = cv.imread('images/dahyun.jpg')
blur = cv.blur(img,(5,5))
blur1 = cv.GaussianBlur(img,(3,3),0)
plot = False
if plot:
    plt.subplot(131),plt.imshow(img),plt.title('Original')
    plt.xticks([]), plt.yticks([])
    plt.subplot(132),plt.imshow(blur),plt.title('Blurred')
    plt.xticks([]), plt.yticks([])
    plt.subplot(133),plt.imshow(blur1),plt.title('Gaussian Blurred')
    plt.xticks([]), plt.yticks([])
    plt.show()
else:
    cv.imshow("Original", img)
    cv.imshow("Blurred", blur)
    cv.imshow("Gaussian Blur", blur1)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

## Median Blur
**Median Blur**
 * highly effective in salt and pepper noise

```python
median_blur = cv.medianBlur(img, 5) #5 is k-size 
cv.imshow("Median Blur", median_blur)
cv.waitKey(0)
cv.destroyAllWindows()
```

## Bilateral Blur
**Bilateral Blur**
* highly effective in noise removal but still keeping the edges

```python
blur = cv.bilateralFilter(img,9,75,75)
blur1 = cv.GaussianBlur(img,(5,5),0)
cv.imshow("Original", img)
cv.imshow("Bilateral Blur", blur)
cv.imshow("Gaussian Blur", blur1)
cv.waitKey(0)
cv.destroyAllWindows()
```
